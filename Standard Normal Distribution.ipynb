### Difference between Standard Normal Distribution and Normal distribution

The normal distribution can have any mean and standard deviation, while the standard normal distribution has a fixed mean of 0 and a fixed standard deviation of 1. The standard normal distribution is a standardized version of the normal distribution,

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
np.random.seed(42) 
data = np.random.normal(0, 1, 1000)  # Mean=0, StdDev=1
x = np.linspace(-4, 4, 1000)
# Compute the PDF values for the standard normal distribution
pdf = norm.pdf(x)
plt.hist(data, bins=30, density=True, alpha=0.6, color='b', edgecolor='black', label='Histogram')
plt.plot(x, pdf, 'r-', lw=2, label='Standard Normal PDF')

plt.xlabel('Value (Z-score)')
plt.ylabel('Probability Density')
plt.title('Histogram and Standard Normal Distribution (Z-distribution)')
plt.legend()

plt.show()
```

## Applications 

```python
1) Z-Scores and Standardization:
    a) Calculating Z-scores (standard scores) to standardize data, making it easier to compare data from different sources or populations.
    b) Z-scores represent the number of standard deviations a data point is away from the mean and can be used to assess how unusual or typical a data point is within a distribution.

2)Hypothesis Testing:
    a) Performing hypothesis tests, such as Z-tests or one-sample Z-tests, to determine if a sample statistic is significantly different from a known population parameter when the population follows a normal distribution.
```
