### Difference between Standard Normal Distribution and Normal distribution

The normal distribution can have any mean and standard deviation, while the standard normal distribution has a fixed mean of 0 and a fixed standard deviation of 1. The standard normal distribution is a standardized version of the normal distribution,

```python
import numpy as np

#During normal distribution mean and sd is user input
mean = 0
std_deviation = 1
sample_size = 100 

samples = np.random.normal(mean, std_deviation, sample_size)
print(samples)
```

```python
import matplotlib.pyplot as plt
plt.hist(samples, bins=30,density=True, alpha=0.6, color='b', edgecolor='black')

plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.title('Histogram of Random Samples from Normal Distribution')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

x = np.linspace(-4, 4, 1000)

# Compute the PDF values for the normal distribution
pdf = norm.pdf(x, mean, std_deviation)
plt.hist(samples, bins=30, density=True, alpha=0.6, color='b', edgecolor='black', label='Histogram')
plt.plot(x, pdf, 'r-', lw=2, label='PDF')

plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.title('Histogram and PDF of Random Samples from Normal Distribution')
plt.legend()
plt.show()
```
