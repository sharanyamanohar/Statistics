5 Number summary gives us a way to describe the distribution using only 5 numbers
1)Minimum
2)1st quantile
3)Median
4)3rd quantile
5)Maximum

```python
import numpy as np
import matplotlib.pyplot as plt

data = np.array([10, 15, 20, 25, 30, 35, 40, 45, 50, 55, 1000])  

minimum = np.min(data)
q1 = np.percentile(data, 25)
median = np.median(data)
q3 = np.percentile(data, 75)
maximum = np.max(data)

iqr = q3 - q1

lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr

plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.boxplot(data, vert=False, labels=['Data'], showfliers=True)
plt.title('Box Plot (Before Removing Outliers)')

filtered_data = data[(data >= lower_bound) & (data <= upper_bound)]
plt.subplot(1, 2, 2)
plt.boxplot(filtered_data, vert=False, labels=['Filtered Data'], showfliers=False)
plt.title('Box Plot (After Removing Outliers)')

plt.tight_layout()
plt.show()
```
