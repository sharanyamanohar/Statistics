Different types of Distribution gives different description of data

## Uniform distribution

Uniform distribution is a probability distribution in which all outcomes are equally likely.eg by tossing a dice we get the probability as 1/6

```python
import numpy as np
import matplotlib.pyplot as plt
data_uniform = np.random.uniform(low=0, high=1, size=1000)

plt.hist(data_uniform, bins=20, density=True, alpha=0.7, color='skyblue', edgecolor='black')

plt.title('Uniform Distribution')
plt.xlabel('Value')
plt.ylabel('Probability Density')

plt.show()
```

## Log-Normal distribution 

```python
Logarithm is normally distributed data.
```

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(2)

mean_log = 0
sigma_log = 0.1  
size = 1000
data_log_normal = np.random.lognormal(mean_log, sigma_log, size)
plt.hist(data_log_normal, bins=30, density=True, alpha=0.7, color='orange', edgecolor='black')

plt.title('Log-Normal Distribution')
plt.xlabel('Value (log scale)')
plt.ylabel('Probability Density')

plt.show()
```

## Pareto distribution 

Pareto distribution is power law distribution,The Pareto principle or "80-20 rule" stating that 80% of outcomes are due to 20% of causes.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import pareto

np.random.seed(42)
alpha = 3.0  
size = 1000

data_pareto = pareto.rvs(alpha, size=size)

plt.hist(data_pareto, bins=30, density=True, alpha=0.7, color='green', edgecolor='black')

plt.title('Pareto Distribution')
plt.xlabel('Value')
plt.ylabel('Probability Density')

plt.show()
```
